```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
bollywood_data = pd.read_csv('Bollywood_filtered_datastes.csv')
bollywood_data
```

```python
gens = set()
for i in bollywood_data['genres']:
    for j in i.split('|'):
        gens.add(j)
gens
```

```python
actors = set()
for i in bollywood_data['actors']:
    for j in str(i).split('|'):
        actors.add(j)
actors
```

```python
conv_dict = {k:v for v, k in dict(enumerate(gens)).items()}
conv_dict
```

```python
conv_dict1 = {k:v for v, k in dict(enumerate(actors)).items()}
conv_dict1
```

```python
lis = []
for i in bollywood_data['genres']:
    tup = []
    for j in tuple(i.split('|')):
        tup.append(conv_dict[j])
    lis.append(tup)
bollywood_data['genres_num'] = lis
```
